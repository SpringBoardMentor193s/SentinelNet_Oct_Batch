```python
import pandas as pd
```

```python
df = pd.read_csv('tennis.csv')
print(df.head())
print(df.tail())
```

```python
print(df.info())
```

```python
print("\n Shape of DataFrame:")
print(df.shape)

print("\n Column Name:")
print(df.columns.tolist())

print("\n Data Types of Each Column:")
print(df.dtypes)

print("\n Descriptive statistics:")
print(df.describe())
```

```python
# single column
print(df['humidity'])
```

```python
#multi column
print(df[['outlook','temp']])
```

```python
print(df.isnull())
print(df.isnull().sum())
df.dropna()
df.fillna(0)
```

```python
#return first row
print('\n first row')
print(df.iloc[0],'\n')

#return first 5 row
print('\n first five row')
print(df.iloc[0:5], '\n')

# returns row by label
print(df.loc[0],'\n') 

# returns column by label
print(df.loc[:, 'outlook'],'\n')      
```
