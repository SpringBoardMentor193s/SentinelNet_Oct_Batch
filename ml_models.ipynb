```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming x and y are defined elsewhere and X_train_scaled and x_test_scaled are created after scaling
# Example placeholders if x and y are not defined:
# import numpy as np
# x = np.random.rand(100, 10)
# y = np.random.randint(0, 2, 100)
# X_train_scaled, x_test_scaled = np.random.rand(80, 10), np.random.rand(20, 10)


x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42, stratify=y)


models = {
    "Logistic Regression": LogisticRegression(random_state=42),
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(),
    "Gradient Boosting": GradientBoostingClassifier()
}

accuracy_scores = {}
y_preds = {}

for name, model in models.items():
    print(f"training {name}...")
    # Using placeholder scaled data - replace with actual scaled data variables if available
    # model.fit(X_train_scaled, y_train)
    # For demonstration purposes, using x_train for now - remove if scaled data is available
    model.fit(x_train, y_train)
    print(f"{name} trained.")

for name, model in models.items():
    print(f"Evaluating {name}...")
    # Using placeholder scaled data - replace with actual scaled data variables if available
    # y_pred = model.predict(x_test_scaled)
    # For demonstration purposes, using x_test for now - remove if scaled data is available
    y_pred = model.predict(x_test)
    y_preds[name] = y_pred

    accuracy = accuracy_score(y_test, y_pred)
    accuracy_scores[name] = accuracy
    print(f"{name} Accuracy: {accuracy * 100:.2f}%")

# Find the best performing model
best_model_name = max(accuracy_scores, key=accuracy_scores.get)
best_y_pred = y_preds[best_model_name]

print(f"\nBest performing model: {best_model_name}")

# Plot confusion matrix for the best performing model
cm = confusion_matrix(y_test, best_y_pred)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=["Normal", "Attack"], yticklabels=["Normal", "Attack"])

plt.title(f'Confusion Matrix for {best_model_name}', fontsize=16)
plt.xlabel('Predicted Label', fontsize=12)
plt.ylabel('True Label', fontsize=12)
plt.show()
```
