```python
import pandas as pd

#Loading Train dataset
train_df = pd.read_csv("kdd_train.csv")

#deleting the "land" column
train_df.drop(['land'], axis = 1, inplace = True)

#Loading Test dataset
test_df = pd.read_csv("kdd_test.csv")

#deleting the "land" column
test_df.drop(['land'], axis = 1, inplace = True)

print("Training Data")
print(f"Shape: {train_df.shape}")
print(train_df.head())

print("\n--- Testing Data ---")
print(f"Shape: {test_df.shape}")
print(test_df.head())
```

```python
# Creating a binary classification for column 'labels' for training dataset
train_df['attack'] = train_df['labels'].apply(lambda x: 0 if x == 'normal' else 1)
```

```python
# Creating a binary classification for column 'labels' for testing dataset
test_df['attack'] = test_df['labels'].apply(lambda x: 0 if x == 'normal' else 1)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6,4))
sns.countplot(x='attack', data=train_df, palette='Set1')
plt.title('Binary Attack Distribution')
plt.xlabel('0 = Normal, 1 = Attack')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10,6))
train_df['labels'].value_counts().head(10).plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Top 10 Attack Types', fontsize=16, fontweight='bold')
plt.xlabel('Attack Type')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.histplot(train_df['duration'], bins=50, log_scale=True, color='teal')
plt.title('Distribution of Connection Duration (Log Scale)', fontsize=15, fontweight='bold')
plt.xlabel('Duration (log scale)')
plt.show()
```

```python
#Fillling all the missing value of mode of the data
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='most_frequent')
df_imputed_array = imputer.fit_transform(train_df)
df_imputed = pd.DataFrame(
    df_imputed_array,
    columns=train_df.columns
)
print("--- Missing values count after imputation ---")
print(df_imputed.isnull().sum().max())
```

```python
#categorical columns
categorical_cols = ['protocol_type', 'service', 'flag','labels']

#one-hot encoding
train_encoded = pd.get_dummies(train_df, columns=categorical_cols)
test_encoded = pd.get_dummies(test_df, columns=categorical_cols)

train_encoded, test_encoded = train_encoded.align(test_encoded, join='inner', axis=1)
print(train_encoded)

print("Encoding Complete")
print(f"Train Encoded Shape: {train_encoded.shape}")
print(f"Test Encoded Shape: {test_encoded.shape}")
```
